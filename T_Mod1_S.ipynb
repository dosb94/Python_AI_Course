```python
#Crea un array de 10x10 con valores aleatorios entre 0 y 1, incluyendo ambos
import numpy as np
array_10x10 = np.random.rand(10, 10)
print(array_10x10)
```

```python


# Convertir a 1D para los cálculos
datos = array_10x10.flatten()

# Cálculos
media = np.mean(datos)
mediana = np.median(datos)
desviacion_std = np.std(datos)

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviación estándar: {desviacion_std:.2f}")
print(f"Datos totales: {len(datos)}")
```

```python
#Resolver sistema de ecuaciones
# Matriz de coeficientes
A = np.array([[3, 5, 2],
              [5, 1, 1],
              [1, -1, 5]])

# Vector de términos independientes
B = np.array([15, 8, 4])

# Resolver el sistema
solucion = np.linalg.solve(A, B)

# Mostrar resultados
print("Solución del sistema:")
print(f"x = {solucion[0]:.2f}")
print(f"y = {solucion[1]:.2f}")
print(f"z = {solucion[2]:.2f}")

# Verificar la solución
print("\nVerificación:")
print(f"3x + 5y + 2z = {3*solucion[0] + 5*solucion[1] + 2*solucion[2]:.2f} = 15")
print(f"5x + y + z = {5*solucion[0] + solucion[1] + solucion[2]:.2f} = 8")
print(f"x - y + 5z = {solucion[0] - solucion[1] + 5*solucion[2]:.2f} = 4")
```

```python
#crear un grafico de la funcion coseno
import matplotlib.pyplot as plt

# Crear datos para el gráfico
x = np.linspace(0, 2*np.pi, 100)  # 100 puntos entre 0 y 2π
y = np.cos(x)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', linewidth=2, label='cos(x)')  # 'r-' = línea roja sólida

# Personalizar el gráfico
plt.title('Función Coseno', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('cos(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)  # Línea horizontal en y=0
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)  # Línea vertical en x=0

# Mostrar el gráfico
plt.show()
```

```python
#Generar un histogrma de datos aleatorios normalmente distribuidos

# Generar 1000 datos aleatorios con distribución normal
# media=0, desviación estándar=1
datos = np.random.normal(0, 1, 1000)

# Crear histograma
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, color='skyblue', edgecolor='black')

plt.title('Histograma - Distribución Normal', fontsize=14)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#Crea un grafico de dispersion con linea de tendencia
# Generar datos de ejemplo
np.random.seed(42)
x = np.random.rand(50) * 10  # 50 puntos aleatorios entre 0 y 10
y = 2 * x + 1 + np.random.randn(50) * 2  # Relación lineal con ruido

# Calcular línea de tendencia (regresión lineal)
coeficientes = np.polyfit(x, y, 1)  # 1 = grado 1 (lineal)
pendiente = coeficientes[0]
interseccion = coeficientes[1]
print(f"Ecuación: y = {pendiente:.2f}x + {interseccion:.2f}")

# Crear puntos para la línea de tendencia
x_tendencia = np.linspace(min(x), max(x), 100)
y_tendencia = pendiente * x_tendencia + interseccion

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6, label='Datos', s=50)
plt.plot(x_tendencia, y_tendencia, color='red', linewidth=2, label=f'Tendencia: y = {pendiente:.2f}x + {interseccion:.2f}')

plt.title('Gráfico de Dispersión con Línea de Tendencia', fontsize=14)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Combina los graficos anteriores en una sola figura
# Generar datos
x = np.random.randn(100) * 10 + 50
y = 2 * x + 10 + np.random.randn(100) * 5

# Crear figura
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histograma
ax1.hist(x, bins=20, color='skyblue', edgecolor='black')
ax1.set_title('Histograma')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

# Dispersión con tendencia
ax2.scatter(x, y, color='blue', alpha=0.5)
ax2.plot(np.sort(x), np.polyval(np.polyfit(x, y, 1), np.sort(x)), 'r-', linewidth=2)
ax2.set_title('Dispersión')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.tight_layout()
plt.show()
```
