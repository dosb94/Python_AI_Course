```python
#Crea una funcion que calcule el IMC
peso = float ( input("Cuanto pesa usted ?"))
estat = float ( input("Cual es su estatura ?"))
IMC = 0

IMC = peso / (estat ** 2 )


if IMC > 29:
    print("Estas gordito")
elif IMC >= 24:  # 24 <= IMC <= 29
    print("Estas bien")
else:  # IMC < 24
    print("Estas bien... flaco")

print(f"Su IMC es: {IMC}")
```

```python
# Crear una funcion que determine si el anio es bisiesto
def bisiesto(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return f"El año {year} es bisiesto"
    else:
        return f"El año {year} no es bisiesto"

year = int(input("Anio a calcular: "))
resultado = bisiesto(year)
print(resultado)
```

```python
# Implementar una funcion que convierta numeros decimales a binarios
def decimales_binarios(num):
    if num == 0:
        return "0"

    binario = ""
    temp = num

    while temp > 0:
        binario = str(temp % 2) + binario
        temp = temp // 2

    return binario

# Pedir el número y mostrar resultado
num = int(input("Introduzca el número: "))
resultado = decimales_binarios(num)
print(f"El número {num} en binario es: {resultado}")
```

```python
# Implementa una funcion que calcule la media de una lista de numeros
def media_lista():
    # Pedir números al usuario
    entrada = input("Ingresa números separados por comas: ")

    # Convertir a lista de floats
    try:
        lista = [float(x.strip()) for x in entrada.split(",")]
    except ValueError:
        print("Error: Ingresa solo números separados por comas")
        return None

    # Calcular y mostrar la media
    if lista:
        media = sum(lista) / len(lista)
        print(f"Lista: {lista}")
        print(f"Media: {media:.2f}")
        return media
    else:
        print("No se ingresaron números")
        return None

# funcion
media_lista()
```
