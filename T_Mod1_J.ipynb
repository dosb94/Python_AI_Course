```python
#Crea una clase rectangulo con metodos de area y perimetro.
def calcular_area(lado1, lado2, lado3):
    # Fórmula de Herón para área de triángulo
    s = (lado1 + lado2 + lado3) / 2
    area = (s * (s - lado1) * (s - lado2) * (s - lado3)) ** 0.5
    return area

def calcular_perimetro(lado1, lado2, lado3):
    return lado1 + lado2 + lado3

# Entrada de datos
l1 = float(input("Dame Lado 1: "))
l2 = float(input("Dame Lado 2: "))
l3 = float(input("Dame Lado 3: "))

# Calcular resultados
area = calcular_area(l1, l2, l3)
perimetro = calcular_perimetro(l1, l2, l3)

# Mostrar resultados con 2 decimales
print(f"Área: {area:.2f}")
print(f"Perímetro: {perimetro:.2f}")
```

```python
#Implementa el método EstadoDeCuenta para para la clase CuentaBancaria
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self.saldo = saldo_inicial

    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto
            print(f"Depósito exitoso: ${monto:.2f}")
        else:
            print("El monto debe ser positivo")

    def retirar(self, monto):
        if monto > 0 and monto <= self.saldo:
            self.saldo -= monto
            print(f"Retiro exitoso: ${monto:.2f}")
        else:
            print("Fondos insuficientes o monto inválido")

    def EstadoDeCuenta(self):
        print(f"\n{'='*30}")
        print("ESTADO DE CUENTA")
        print(f"{'='*30}")
        print(f"Titular: {self.titular}")
        print(f"Saldo disponible: ${self.saldo:.2f}")
        print(f"{'='*30}")

# Uso de la clase
mi_cuenta = CuentaBancaria("Ana García", 500.75)
mi_cuenta.EstadoDeCuenta()

mi_cuenta.depositar(200)
mi_cuenta.retirar(100)
mi_cuenta.EstadoDeCuenta()
```

```python
#Implementa una clase padre Vehículo y una clase hija Moto, con atributos y métodos apropiados
class Vehiculo:
    def __init__(self, marca, modelo, color, año):
        # 5 cosas que tienen en común
        self.marca = marca      # 1. Marca
        self.modelo = modelo    # 2. Modelo
        self.color = color      # 3. Color
        self.año = año          # 4. Año de fabricación
        self.encendido = False  # 5. Estado encendido/apagado

    def encender(self):
        self.encendido = True
        print(f"{self.marca} {self.modelo} encendido")

    def apagar(self):
        self.encendido = False
        print(f"{self.marca} {self.modelo} apagado")

    def mostrar_info(self):
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Color: {self.color}")
        print(f"Año: {self.año}")
        print(f"Estado: {'Encendido' if self.encendido else 'Apagado'}")


class Moto(Vehiculo):
    def __init__(self, marca, modelo, color, año, cilindrada, tipo_manillar):
        super().__init__(marca, modelo, color, año)
        # Atributos específicos de moto
        self.cilindrada = cilindrada  # en cc
        self.tipo_manillar = tipo_manillar  # deportivo, cruiser, etc.
        self.caballito = False  # estado del caballito

    def hacer_caballito(self):
        if self.encendido:
            self.caballito = True
            print("¡Haciendo caballito!")
        else:
            print("Primero enciende la moto")

    def bajar_caballito(self):
        self.caballito = False
        print("Moto estabilizada")

    def mostrar_info(self):
        super().mostrar_info()  # llama al método del padre
        print(f"Cilindrada: {self.cilindrada}cc")
        print(f"Tipo de manillar: {self.tipo_manillar}")
        print(f"Caballito: {'Sí' if self.caballito else 'No'}")


# Ejemplo de uso
print("=== VEHÍCULO GENÉRICO ===")
carro = Vehiculo("Toyota", "Corolla", "Rojo", 2020)
carro.encender()
carro.mostrar_info()

print("\n=== MOTO ===")
mi_moto = Moto("Yamaha", "MT-07", "Azul", 2022, 689, "Deportivo")
mi_moto.encender()
mi_moto.hacer_caballito()
mi_moto.mostrar_info()
```

```python
#Implementa una clase Libro con atributos y métodos apropiados
class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponible = True

    def prestar(self):
        if self.disponible:
            self.disponible = False
            return f"'{self.titulo}' ha sido prestado"
        return f"'{self.titulo}' no está disponible"

    def devolver(self):
        self.disponible = True
        return f"'{self.titulo}' ha sido devuelto"

    def __str__(self):
        return f"'{self.titulo}' por {self.autor} - {'Disponible' if self.disponible else 'Prestado'}"

# Uso simple
libro = Libro("Don Quijote", "Miguel de Cervantes")
print(libro)
print(libro.prestar())
print(libro)
print(libro.devolver())
print(libro)
```
